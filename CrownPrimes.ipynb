Divisibility for sorting a bunch of crowns of various sizes

```python
#Generate all partitions of the list L
def allPartitions(L):
    if len(L) == 1:
        return [(tuple(L),)]
    else:
        ret = []
        parts = allPartitions(L[1:])
        for part in parts:
            ret.append(((L[0],),*part))
            for i in range(len(part)):
                p = []
                for j in range(len(part)):
                    if i == j:
                        p.append((L[0],) + part[j])
                    else:
                        p.append(part[j])
                ret.append(tuple(p))
    return ret
```

```python
factors = {1:{},2:{2:1},3:{3:1}}
def factor_integers(N):
    global factors
    for n in range(4, N+1):
        if n not in factors:
            factors[n] = dict()
            k = n
            i = 2
            while True:
                if k % i == 0:
                    factors[n][i] = 0
                    while k % i == 0:
                        factors[n][i] += 1
                        k //= i
                if k == 1:
                    break
                i += 1
                if k == n and i**2 > n:
                    factors[n][n] = 1
                    break
```

```python
def crownSortCount(partition):
    global factors
    d = 0
    g = []
    p = []
    c = []
    for part in partition:
        d += sum(part)
        g.append(sum(part))
        p.append(len(part))
        c += part
    k = len(partition)
    q = len(c)
    d += q
    factor_integers(d)
    ret = dict()
    for i in range(2, d+1):
        for j in factors[i]:
            if j not in ret:
                ret[j] = 0
            ret[j] += factors[i][j]
    ret[2] += d + q - 3*k
    for i in range(q):
        for j in factors[c[i]]:
            ret[j] += factors[c[i]][j]
    for j in range(k):
        for l in range(p[j] - 1):
            for i in factors[g[j] + l]:
                ret[i] += factors[g[j] + l][i]
        for l in range(2, g[j] + p[j] + 1):
            for i in factors[l]:
                ret[i] -= factors[l][i]
                if ret[i] == 0:
                    del ret[i]
    return ret
```

```python
crownSortCount([[2,3,9],[7,8]])
```

```python
def crownSortCounts(crowns):
    partitions = allPartitions(crowns)
    ret = dict()
    for partition in partitions:
        ret[partition] = crownSortCount(partition)
    return ret
```

```python
csc = crownSortCounts([2,3,7,8,9])
csc
```
