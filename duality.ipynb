```python
g = MPScount2.AdjacencyGraphInfo()
g.addW(3)
g.addM(2)
gg = MPSduality.AdjacencyGraph(g)
mpsList = MPSduality.getAllMPS(gg)
mpsRev = MPSduality.getReversals(gg, mpsList)
print(len(mpsRev))
pairs = [(mpsList[i], mpsRev[i]) for i in range(len(mpsList))]
outfile = "MW23.txt"
with open(outfile, 'w') as out:
    for pair in pairs:
        out.write(str(pair[0]))
        out.write('\n')
        out.write(str(pair[1]))
        out.write('\n\n')
```

```python
g = MPScount2.AdjacencyGraphInfo()
g.addW(3)
g.addM(2)
gg = MPSduality.AdjacencyGraph(g, shuffle = True)
mpsList = MPSduality.getAllMPS(gg)
mpsRev = MPSduality.getReversals(gg, mpsList)
print(len(mpsRev))
pairs = [(mpsList[i], mpsRev[i]) for i in range(len(mpsList))]
outfile = "MW23shuffle.txt"
with open(outfile, 'w') as out:
    for pair in pairs:
        out.write(str(pair[0]))
        out.write('\n')
        out.write(str(pair[1]))
        out.write('\n\n')
```

```python
#Look at just the ones that start with a left-W CJ and later CJ the M and W together
for pair in pairs:
    if pair[0][0].op == "CJ" and pair[0][0].left1 == True:
        for op in pair[0][1:]:
            if op.op == "CJ" and op.type1 == MPSduality.M:
                print(pair[0])
                print(pair[1])
                print()
                break
```

```python
#Look at just the ones that start with a right-W CJ and later CJ the M and W together
for pair in pairs:
    if pair[0][0].op == "CJ" and pair[0][0].left1 == False:
        for op in pair[0][1:]:
            if op.op == "CJ" and op.type1 == MPSduality.M:
                print(pair[0])
                print(pair[1])
                print()
                break
```
